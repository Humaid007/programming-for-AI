```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.preprocessing import LabelEncoder
import joblib
```

```python
train_df = pd.read_csv(r"C:\Users\pc\Downloads\Training.csv")
test_df = pd.read_csv(r"C:\Users\pc\Downloads\Testing.csv")
```

```python
train_df = train_df.drop(columns=['Unnamed: 133'], errors='ignore')
test_df = test_df.drop(columns=['Unnamed: 133'], errors='ignore')
```

```python
X_train = train_df.drop('prognosis', axis=1)
y_train = train_df['prognosis']
```

```python
X_test = test_df.drop('prognosis', axis=1)
y_test = test_df['prognosis']
```

```python
le = LabelEncoder()
y_train_encoded = le.fit_transform(y_train)
y_test_encoded = le.transform(y_test)
```

```python
train_df.head()
```

```python
train_df.tail()
```

```python
model = RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(X_train, y_train_encoded)
```

```python
y_pred = model.predict(X_test)
```

```python
print("Accuracy:", accuracy_score(y_test_encoded, y_pred))
print("\nClassification Report:\n", classification_report(y_test_encoded, y_pred, target_names=le.classes_))
print("\nConfusion Matrix:\n", confusion_matrix(y_test_encoded, y_pred))
```

```python
joblib.dump(model, 'medical_diagnosis_model.pkl')
joblib.dump(le, 'label_encoder.pkl')
```
